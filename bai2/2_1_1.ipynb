#### TASK: Vẽ một hình chữ nhật trống MÀU ĐỎ quanh khuôn mặt của chú chó và hiển thị trong notebook.

```python
import cv2
from PIL import ImageDraw


image_cv2 = cv2.imread(image_path)
gray = cv2.cvtColor(image_cv2, cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

image_pil = Image.open(image_path)
draw = ImageDraw.Draw(image_pil)

for (x, y, w, h) in faces:
    draw.rectangle([x, y, x + w, y + h], outline="red")

plt.imshow(image_pil)
plt.show()
```

#### TASK: Vẽ một HÌNH TAM GIÁC MÀU XANH DƯƠNG ở giữa ảnh. Có thể chọn kích thước và góc tùy ý, nhưng phải là hình tam giác (3 cạnh) theo bất kỳ hướng nào.

```python
image = Image.open(image_path)

draw = ImageDraw.Draw(image)

point1 = (image.width // 2, image.height // 2 - 50)
point2 = (image.width // 2 - 50, image.height // 2 + 50)
point3 = (image.width // 2 + 50, image.height // 2 + 50)

# Vẽ tam giác bằng 3 đoạn thẳng
draw.line([point1, point2], fill="blue", width=3)
draw.line([point2, point3], fill="blue", width=3)
draw.line([point3, point1], fill="blue", width=3)

plt.imshow(image)
plt.show()
```

### BONUS TASK. Bạn có thể tô màu tam giác này không? Điều này cần một hàm khác, không có trong bài giảng. Hãy thử xem liệu bạn có thể tìm thấy trên Google không.

[CLICK VÀO LIÊN KẾT SAU ĐỂ XEM GỢI Ý](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
triangle_points = [point1, point2, point3]
draw.polygon(triangle_points, fill="blue")
plt.imshow(image)
plt.show()
```
